```python
# Check misdecoded UTF-8
def check_fin_letter_encoding(file_path):
    with open(file_path, "rb") as f:
        raw = f.read()
        num_a =raw.count(b'\xc3\xa4')
        num_o =raw.count(b'\xc3\xb6')
        num_u =raw.count(b'\xc3\xbc')
    print(f"Number of 'ä': {num_a}, 'ö': {num_o}, 'ü': {num_u}")    
```

```python
def check_encoding(file_path):
    with open(file_path, "rb") as f:
        raw = f.read()

    # 1. Is the file valid UTF-8?
    try:
        raw.decode("utf-8")
        print("File is valid UTF-8")
    except UnicodeDecodeError as e:
        print("Invalid UTF-8:", e)

    # 2. Check how many UTF-8 ä bytes exist
    print("UTF-8 ä bytes:", raw.count(b'\xc3\xa4'))
```

```python
# Repair misdecoded UTF-8
from ftfy import fix_text
def fix_encoding(input_file, output_file):
    print(f"Fixing encoding for file: {input_file}")
    with open(input_file, "r", encoding="utf-8") as f:
        content = f.read()

    fixed = fix_text(content)

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(fixed)  
    print(f"Saved fixed file: {output_file}")    
```
